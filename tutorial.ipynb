```python
import numpy as np # this libraries allows further arithmetic operations
pows=np.linspace(0.5,2.,5);sd=1e6;
L=10**(pows)
mo=(16/7)*sd*L**3;mag=(2/3)*(np.log10(mo)-9.1)
print(pows,"\n",L,"\n",mag)
```

```python
fig,[ax1,ax2]=plt.subplots(ncols=1,nrows=2,figsize=(4,5))
ax1.plot(L,mag,'o');
ax1.plot(L,mo,'o');
ax2.loglog(L,mag,'o');
ax2.loglog(L,mo,'o');
```

```python
from pylab import xlabel, ylabel

def lolo(xx,yy,yy2,labx,laby,laby2):
    fig,ax=plt.subplots(figsize=(6,8))
    ax.semilogx(xx,yy,'r')
    xlabel(labx); ylabel(laby)
    ax.xaxis.grid(True,which='both',alpha=1)
    ax2 = ax.twinx()
    ax2.loglog(xx, yy2, 'b--')
    ax2.set_ylabel(laby2, color='r')
    ax2.tick_params('y', colors='r')
    ax2.yaxis.grid(True,which='both',alpha=1)
```

```python
pows=np.linspace(0.5,2.,5);L=10**(pows)
sd=5e6;mo=(16/7)*sd*L**3;mag=(2/3)*(np.log10(mo)-9.1)
# LeTeX support for matplotlib, with pslatex package :
import matplotlib
plt.rc('text', usetex=True);plt.rc('font', family='serif')
matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}',r'\usepackage{pslatex}']
# call the plotting function lolo:
lolo(L,mag,mo,'Length L (m)','M$_w$','M$_o$ (N m)')
```
